# Conv2D

https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html

It is called 2 dimensional CNN because the kernel slides along 2 dimensions on the data

padding  – Zero-padding added to both sides of the input. Default: 0

## Output size

ouput size = [((input size + 2 * (padding) - covolution filter size)/stride) + 1] X [((input size + 2 * (padding) - covolution filter size)/stride) + 1]

```python
input_size_x = 6
input_size_y = 6
padding_size = 1
filter_size = 3
stride_size = 1

ouput_size_height_x = (((input_size_x + 2 * (padding_size) - filter_size)/stride_size) + 1)
ouput_size_widtht_y = (((input_size_y + 2 * (padding_size) - filter_size)/stride_size) + 1)

print("Output size ---> height * width (X x Y) = ", ouput_size_height_x, " x ", ouput_size_widtht_y )
```

# Convolutional module

in_channels – Number of channels in the input image

out_channels – Number of channels produced by the convolution // Number of filters

kernel_size – Size of the convolving kernel // filter size

stride – Stride of the convolution. Default: 1

padding – Zero-padding added to both sides of the input. Default: "0". Add "1" to produce the output with original
          image size.
          
* The parameters kernel_size, stride, padding can either be:
      a single int – in which case the same value is used for the height and width dimension
      a tuple of two ints – in which case, the first int is used for the height dimension, and the second int for the width dimension

```python
mod2 = nn.Conv2d( in_channels = 2 ,out_channels= 5 ,  kernel_size=3,  padding=1, stride = 1, bias = True )
print(mod2)
```

# Building a Convolutional Neural Network

### Strategy

    1. At each convolution layer - double the feature number (in_channels to out_channels (i.e 2 times of in_channels)) at each convolution layer
    2. At each maxpooling layer - reduce the size of the image by 2 // Also useful in creating global invariance
    
Note: Trading image resolution with features number. Controlling the computational complexity. (Image size/2 and #features X 2)

The convolutional and maxpooling layers are used for feature extraction. Linear layers are used for classification.

Reshape i.e. Flatten the extracted features as (batch size, number of features) when passing through the linear layer 

```python
class LeNet5_convnet(nn.Module):

    def __init__(self):

        super(LeNet5_convnet, self).__init__()
        self.convlayer1 = nn.Conv2d( in_channels = 1 ,out_channels= 50 ,  kernel_size=3,  padding=1, stride = 1, bias = True )
        self.poolmax1 = nn. MaxPool2d(kernel_size =(2,2))
        self.convlayer2 = nn.Conv2d( in_channels = 50 ,out_channels= 100 ,  kernel_size=3,  padding=1, stride = 1, bias = True )
        self.poolmax2 = nn. MaxPool2d(kernel_size =(2,2))
        self.linearlay1 = nn.Linear(in_features= 4900, out_features= 100, bias= True)
        self.linearlay2 = nn.Linear(in_features= 100, out_features= 10, bias= True)
        
        
    def forward(self, x):
        x = self.convlayer1(x)
        x = torch.relu(x)
        x = self.poolmax1(x)
        x = self.convlayer2(x)
        x = torch.relu(x)
        x = self.poolmax2(x)
        x = self.linearlay1(x.view(-1,4900))
        x =torch.relu(x)
        scores = self.linearlay2(x)
        
        return scores
        
convnet = LeNet5_convnet()
print(convnet)
        
```
